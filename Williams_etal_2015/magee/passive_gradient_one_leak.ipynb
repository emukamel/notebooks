A simple multi-compartment extension of [O'Leary et al. (2014)](http://alexhwilliams.info/pubs/OLeary_etal_2014.pdf)
-------------

This is a model of activity-dependent conductance regulation, based on the time-averaged intracellular calcium concentration. While this can [(and has)](http://alexhwilliams.info/pubs/OLeary_etal_2014.pdf) been used to regulate voltage-dependent conductances with Hodgkin-Huxley-like channel models, a simpler conductance-based model with only leak currents suffices to capture the model's behavior. This is because the model integrates the average calcium concentration over a fairly large time window, and is therefore insensitive to the frequency of calcium transients (but see [Liu et al., 1998](http://neurotheory.columbia.edu/~larry/LiuJNeurosci98.pdf)). Voltage-dependent conductances introduce bifurcations in conductance space, making the time-averaged calcium function discontinuous but this [typically$^1$](#footnote1) does not affect the behavior of the model.

We consider a cable of connected eletrical compartments, with the leftmost compartment representing the soma. The somatic calcium concentration, controls the production of mRNA for each leak conductance. The mRNA is trafficked between neighboring compartments with anterograde rate constants, $a_k$, and retrograde rate constants, $b_k$. Each mRNA is degraded in a compartment specific manner, with rate constant $\beta_k$:

$$
\begin{array}{ccc}
u_1 & \overset{a_1}{\underset{b_1}{\rightleftharpoons}} &
u_2 & \overset{a_2}{\underset{b_2}{\rightleftharpoons}} &
u_3 & \overset{a_3}{\underset{b_3}{\rightleftharpoons}} &
u_4 & \overset{a_4}{\underset{b_4}{\rightleftharpoons}} & ...
\\
\beta_1 \Bigg\downarrow \phantom{\beta_1} & &
\beta_2 \Bigg\downarrow \phantom{\beta_2} & &
\beta_3 \Bigg\downarrow \phantom{\beta_3} & &
\beta_4 \Bigg\downarrow \phantom{\beta_4} & &
\\
\varnothing &  &
\varnothing &  &
\varnothing &  &
\varnothing &  &
\end{array}
$$

The code below numerically simulates the following set of equations.

$$ \dot{V}_k = \sum_i g_{i,k} (E_i - V_k) + \sum_{j \in \mathscr{N}} g_{ax} (V_j - V_k) ~~~~~(1)$$

$$ ~~~~~~~~~~~~~~~~~~~~~ c(V) = \frac{1}{1+\exp(-V/20)}~~~~~~~~~~~~~~~~~(2)$$

$$ ~~~~~~~~~~~~~~~~~~~~~~~~~~~ \dot{g}_{i,k} = \frac{u_{i,k} - g_{i,k}}{\tau_g} ~~~~~~~~~~~~~~~~~~~~~~~~~(3a)$$

$$ ~~~~~~~~~~~~ \dot{u}_{i,1} = \alpha_i(T) - \beta_{i,1} u_{i,1} - a_1 u_{i,1} + b_1 u_{i,2} ~~~~~~~~~~~~~~~~~~~(4a)$$

$$ ~~~~~~ \dot{u}_{i,k} = - \beta_{i,k} u_{i,k} - (a_k + b_{k-1}) u_{i,k} + b_k u_{i,k+1} + a_{k-1} u_{i,k-1} 
~~~~(4b)$$

$$ ~~~~~~~ \dot{T} = \alpha_T(c(V_1)) - \beta_T~~~~~~~~~~~~~~~~~~~~~~(5)$$

Equation (1) is the membrane equation, where $\mathscr{N}$ is the set of neighboring electrical compartments. Equation (2) specifies the intracellular calcium level, which is approximated as a monotonic function of membrane potential. Equation (3a) specifies the change in each leak conductance, $g_i$. Equation (4a) describes the change in mRNA in the somatic compartment (including production based on the calcium-dependent factor $T$); equation (4b) describes the change in mRNA in all other compartments. Equation 3 causes $g_{i,k}$ to approach $u_{i,k}$, the mRNA concentration, over time. We assume this process occurs for all ion channels with the same time constant, $\tau_g$. It is possible to instead solve this equation with channel-specific insertion and removal rate constants:

$$ ~~~~ \frac{dg_i}{dt} = \alpha_{g,i} u_i - \beta_{g,i} g_i~~~~~~~~~~~~~~~~~~~~~~~(3b)$$

However, the simplified form of equation (3a) cuts down the parameters in the model without modifying its overall behavior. Finally, the production of each mRNA is determined by the concentration of a 'master' transcription factor $T$, whose dynamics are provided by equation (5); $\alpha_T$ and $\beta_T$ are constants. Assuming constant (first-order) degradation is necessary for conductance regulation to implement integral control ([O'Leary et al., 2014](http://alexhwilliams.info/pubs/OLeary_etal_2014.pdf); [Drenstig et al., 2008](http://dx.doi.org/10.1021/jp806818c)). This can cause $T$ to go negative. However, I fix the input to $\alpha_i(\cdot)$ to be $\max(0,T)$. Thus, all mRNA concentrations are restricted to be nonnegative.

It is reasonable to model $\alpha_i(T)$ in equation (4) as well as $\alpha_T(c)$ in equation (5) using scaled Hill Equations:

$$ ~~~~ \alpha_i(T) = \alpha_{min} + (\alpha_{max} - \alpha_{min}) \left ( \frac{T^n}{T^n + K_d} \right ) ~~~~~~~~~~~~~~~~~~~~~~~(6a)$$
$$ ~~~ \alpha_i(T) = \alpha_{min} + (\alpha_{max} - \alpha_{min}) \left ( 1-\frac{T^n}{T^n + K_d} \right ) ~~~~~~~~~~~~~~~~~~~~(6b)$$

Equation (6a) describes a scenario where the binding of $T$ induces transcription, while (6b) describes a scenario where $T$ suppresses transcription. In both cases $\alpha_{max}$ and $\alpha_{min}$ are the soft bounds on the transcription rate. Note that $i$ is an index variable for the mRNA/channel type.

Note that the 'target', 'set point', or steady-state calcium concentration, $c_{ss}$, is found by solving:

$$  c_{ss} = \alpha_T^{-1} ( \beta_T ) $$

When $c = c_{ss}$, the system is in equilibrium, since $dT/dt = 0$. Appropriate parameter sets will cause the somatic calcium, $c_1$, to approach $c_{ss}$ over time (as well as homeostatically regulate $c_1$ around this set point in response to perturbations).

```python
from __future__ import division
import numpy as np
import pylab as plt
%matplotlib inline

# Number of compartments and leak currents in the cable
nComp = 8 
nChan = 1

# Parameters for equations 1-3
E = np.array([80])  # list of reversal potentials each leak current (mV)
tau_g = 1.0                  # arbitrary time units
g_ax = 1.0                    # uS/nF
c = lambda V: 1/(1+(np.exp(-(V+40)/20))) # function for calcium, e.g. c(0) yields 0.5

# Equation 4, with alpha given by equations 6a & 6b
beta = 1e-2*np.ones((nComp,nChan))
alpha = []
alpha.append(lambda T: 0.0 + (T/(T+4)))

# Parameters for mRNA trafficking, produces linear gradient for small beta
a,b = [],[]
for i in range(1,nComp):
    b.append(i*np.ones(nChan))
    a.append((i+1)*np.ones(nChan))
a = np.array(a)
b = np.array(b)

# Equation 5, steady state at c = beta_T / alpha_T
alpha_T = lambda ca: 0.1*(1-(ca**2/(ca**2+0.2)))
beta_T = 0.05
```

```python
## This cell plots some useful information about the parameters
g_cols = [(0.89, 0.10, 0.10),
          (0.21, 0.49, 0.72),
          (0.30, 0.68, 0.29)]

v_axis = np.linspace(-80,50,100)
T_axis = np.linspace(0,5,100)

# Calcium as a function of membrane potential
plt.figure()
plt.plot(v_axis,c(v_axis),lw=2)
plt.xlabel('membrane potential (mV)',fontweight='bold')
plt.ylabel('Ca conc. (a.u.)',fontweight='bold')
plt.xlim([v_axis[0],v_axis[-1]])
plt.show()

# Transcription rate as a function of master transcription factor for each channel type
# --> legend labels each channel type by the reversal potential
plt.figure()
for (i,alpha_func) in enumerate(alpha):
    plt.plot(T_axis,alpha_func(T_axis),'-',c=g_cols[i],lw=2,label='E = '+str(E[i]))
plt.xlabel('conc. of T',fontweight='bold')
plt.ylabel('transcription rate',fontweight='bold')
plt.legend(loc='right')
plt.show()

# Transcription factor activation rate as a function of calcium
plt.figure()
plt.plot(c(v_axis),alpha_T(c(v_axis))-beta_T,'-k',lw=2)
plt.plot(c(v_axis),np.zeros(len(v_axis)),'--r',lw=2)
plt.xlabel('Ca conc. (a.u.)',fontweight='bold')
plt.ylabel('dT/dt',fontweight='bold')
target_index = np.argmin(abs(alpha_T(c(v_axis))-beta_T))
ca_targ = c(v_axis[target_index])
plt.xlim([min(c(v_axis)),max(c(v_axis))])
plt.show()
```

```python
## This cell does the simulation
from scipy.integrate import odeint # import ODE solver

# steady-state membrane potential
def v(g):
    return np.sum(g*E)/np.sum(g)

# a rectifier function to prevent T<0
def fix(x):
    if x < 0: return 0
    else: return x

# function that returns dy/dt given vector y as input
def model(y,t0,param):
    # y[:nComp] are the membrane potentials, v
    # y[nComp:nComp+(nComp*nChan)] are the mRNA concentrations, u
    # y[nComp+(nComp*nChan):nComp+(nComp*nChan)*2] are the conductances, g
    # y[-1], the last element, is T
    
    dydt = []  # list of derivatives, vars in same order as y

    v = y[:nComp]
    u = np.reshape(y[nComp:nComp+(nComp*nChan)],(nComp,nChan))
    g = np.reshape(y[nComp+(nComp*nChan):nComp+2*(nComp*nChan)],(nComp,nChan))
    T = fix(y[-1])
    
    c_now = c(v[0])  # current calcium concentration
    
    # Change in membrane potential
    for i in range(nComp):
        dydt.append(np.dot(g[i,:],(E - v[i])) + param[0]*(-80 - v[i]))
        if i > 0:
            dydt[-1] += g_ax*(v[i-1] - v[i])
        if i < (nComp-1):
            dydt[-1] += g_ax*(v[i+1] - v[i])
    
    # Change in mRNA
    for i in range(nComp):
        for (j,alpha_func) in enumerate(alpha):
            dydt.append(-beta[i,j]*u[i,j])
            if i == 0: dydt[-1] += alpha_func(T)                           # production in soma
            if i > 0: dydt[-1] += -b[i-1,j]*u[i,j] + a[i-1,j]*u[i-1,j]     # trafficking to left
            if i < (nComp-1):
                dydt[-1] += -a[i,j]*u[i,j] + b[i,j]*u[i+1,j] # trafficking to right
            
    # Change in conductance
    for i in range(nComp):
        for j in range(nChan):
            dydt.append((u[i,j]-g[i,j])/tau_g) # Equation (3)
    
    # Change in T
    dydt.append(alpha_T(c_now) - beta_T)
    
    # Return calculated derivatives
    return dydt


def run_sim():
    # initial conductances
    gLconst = 1.0*np.random.uniform(0.75,1.5)
    v0 = -60*np.ones(nComp)
    g0 = np.random.uniform(0.1,0.001,nChan*nComp)
    u0 = g0          # initial mRNA
    T0 = [1.0]       # initial transcription factor

    # initial state
    y0 = np.concatenate((v0,g0,u0,T0))
    t = np.linspace(0,1e3,500)

    # Numerically solve the equations given by the function 'model'
    y = odeint(model, y0, t,args=([gLconst],))

    v = y[:,:nComp]
    u = y[:,nComp:nComp+(nComp*nChan)]
    g = y[:,nComp+(nComp*nChan):nComp+2*(nComp*nChan)]
    T = y[:,-1]
    ca = c(v[:,0])
    
    return (t,v,u,g,T,ca)

(t,v,u,g,T,ca) = run_sim()
```

```python
comp_cols = np.array([np.linspace(0.4,0,nComp),np.linspace(0.4,0,nComp),np.linspace(0.4,0.7,nComp)]).T

plt.figure(figsize = (10,10))
plt.subplot(511)
for i in range(nComp): plt.plot(t,v[:,i],c=comp_cols[i],lw=2,label='comp '+str(i))
plt.xticks([])
plt.yticks([-60,-30,0])
plt.ylabel('membrane\npotential (mV)',fontweight='bold')
plt.legend(loc='right')

plt.subplot(512)
for i in range(nComp*nChan): plt.plot(t,u[:,i],c=g_cols[i%nChan])
plt.ylabel('mRNA conc.',fontweight='bold')
plt.xticks([])

plt.subplot(513)
for i in range(nComp*nChan): plt.plot(t,g[:,i],c=g_cols[i%nChan])
plt.xticks([])
plt.ylabel('g (uS/nF)',fontweight='bold')

plt.subplot(514)
plt.plot(t,T,lw=2)
plt.xticks([])
plt.ylabel('T conc.',fontweight='bold')

plt.subplot(515)
plt.plot(t,ca,lw=2)
plt.plot([t[0],t[-1]],np.ones(2)*ca_targ,'--r',lw=2,label='set point')
plt.legend(loc='lower right')
plt.xlabel('time (a.u.)', fontweight='bold')
plt.ylabel('ca conc.', fontweight='bold')

plt.tight_layout()
plt.savefig('model_trace.svg')
plt.show()
```

```python
## Run a series of simulations
import colorsys
N = 10
plt.figure(figsize=(13,5))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

gg = []
h = 0
for k in range(N):
    (t,v,u,g,T,ca) = run_sim()
    gg.append(g[-1,:])
    h += 0.618033988749895 # silver ratio
    ax2.plot(g[-1,:],'o-',color=colorsys.hsv_to_rgb((h%1),0.8,0.95))

plt.sca(ax1)
gg = np.array(gg)
plt.errorbar(range(nComp),np.mean(gg,axis=0),yerr=np.std(gg,axis=0),fmt='o-',color='k')
plt.title('Steady-State conductance gradient\nPopulation Average',fontweight='bold',fontsize=14)
plt.xticks(range(N))
plt.xlabel('Distance from soma',fontweight='bold',fontsize=12)
plt.ylabel('Conductance (uS/nF)',fontweight='bold',fontsize=12)
plt.xlim([-0.5,8.5])

plt.sca(ax2)
plt.title('Steady-State conductance gradient\nSingle-Cell Data',fontweight='bold',fontsize=14)
plt.xticks(range(N))
plt.xlabel('Distance from soma',fontweight='bold',fontsize=12)
plt.ylabel('Conductance (uS/nF)',fontweight='bold',fontsize=12)
plt.xlim([-0.5,8.5])

plt.savefig('model_magee.svg')
plt.show()
```

```python
def run_perturb():
    # initial conductances
    gLconst = 1.0*np.random.uniform(0.75,1.5)
    v0 = -60*np.ones(nComp)
    g0 = np.random.uniform(0.1,0.001,nChan*nComp)
    u0 = g0          # initial mRNA
    T0 = [1.0]       # initial transcription factor

    # initial state
    y0 = np.concatenate((v0,g0,u0,T0))
    t = np.linspace(0,1e3,500)

    # Numerically solve the equations given by the function 'model'
    y = odeint(model, y0, t,args=([gLconst],))

    v1 = y[:,:nComp]
    u1 = y[:,nComp:nComp+(nComp*nChan)]
    g1 = y[:,nComp+(nComp*nChan):nComp+2*(nComp*nChan)]
    T1 = y[:,-1]
    ca1 = c(v1[:,0])
    
    y2 = odeint(model, y[-1,:],t,args=([2*gLconst],))
    
    v2 = y2[:,:nComp]
    u2 = y2[:,nComp:nComp+(nComp*nChan)]
    g2 = y2[:,nComp+(nComp*nChan):nComp+2*(nComp*nChan)]
    T2 = y2[:,-1]
    ca2 = c(v2[:,0])
    
    return (np.append(t,t+t[-1]),
            np.append(v1,v2,axis=0),
            np.append(u1,u2,axis=0),
            np.append(g1,g2,axis=0),
            np.append(T1,T2,axis=0),
            np.append(ca1,ca2,axis=0))
```

```python
(t,v,u,g,T,ca) = run_perturb()

plt.figure(figsize = (10,10))
plt.subplot(511)
for i in range(nComp): plt.plot(t,v[:,i],c=comp_cols[i],lw=2,label='comp '+str(i))
plt.xticks([])
plt.yticks([-60,-30,0])
plt.ylabel('membrane\npotential (mV)',fontweight='bold')
plt.legend(loc='right')

plt.subplot(512)
for i in range(nComp*nChan): plt.plot(t,u[:,i],c=g_cols[i%nChan])
plt.ylabel('mRNA conc.',fontweight='bold')
plt.xticks([])

plt.subplot(513)
for i in range(nComp*nChan): plt.plot(t,g[:,i],c=g_cols[i%nChan])
plt.xticks([])
plt.ylabel('g (uS/nF)',fontweight='bold')

plt.subplot(514)
plt.plot(t,T,lw=2)
plt.xticks([])
plt.ylabel('T conc.',fontweight='bold')

plt.subplot(515)
plt.plot(t,ca,lw=2)
plt.plot([t[0],t[-1]],np.ones(2)*ca_targ,'--r',lw=2,label='set point')
plt.legend(loc='lower right')
plt.xlabel('time (a.u.)', fontweight='bold')
plt.ylabel('ca conc.', fontweight='bold')

plt.tight_layout()
plt.savefig('model_trace_perturb.svg')
plt.show()
```

```python
## Run a series of simulations
import colorsys
N = 10
plt.figure(figsize=(13,5))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

gg1,gg2 = [],[]
h = 0
for k in range(N):
    (t,v,u,g,T,ca) = run_perturb()
    gg1.append(g[len(t)/2,:])
    gg2.append(g[-1,:])
    
    if k < 3:
        ax2.plot(g[len(t)/2,:],'o-',color=colorsys.hsv_to_rgb((h%1),0.8,0.95))
        ax2.plot(g[-1,:],'o-',color=colorsys.hsv_to_rgb((h%1),0.8,0.95))
        h += 0.618033988749895 # silver ratio
        
plt.sca(ax1)
gg1,gg2 = np.array(gg1),np.array(gg2)
plt.errorbar(range(nComp),np.mean(gg1,axis=0),yerr=np.std(gg1,axis=0),fmt='o-',color='k')
plt.errorbar(range(nComp),np.mean(gg2,axis=0),yerr=np.std(gg2,axis=0),fmt='o-',color='k')
plt.title('Steady-State conductance gradient\nPopulation Average',fontweight='bold',fontsize=14)
plt.xticks(range(N))
plt.xlabel('Distance from soma',fontweight='bold',fontsize=12)
plt.ylabel('Conductance (uS/nF)',fontweight='bold',fontsize=12)
plt.xlim([-0.5,8.5])

plt.sca(ax2)
plt.title('Steady-State conductance gradient\nSingle-Cell Data',fontweight='bold',fontsize=14)
plt.xticks(range(N))
plt.xlabel('Distance from soma',fontweight='bold',fontsize=12)
plt.ylabel('Conductance (uS/nF)',fontweight='bold',fontsize=12)
plt.xlim([-0.5,8.5])

plt.savefig('model_magee_perturb.svg')
plt.show()
```

**\[1\]** Including voltage and time-dependent ionic conductances in the model introduces bifurcations in the electrical activity (e.g. the transition from quiescent to tonic spiking), resulting in discontinuities in the time-average calcium signal. This can cause indefinite oscillations or "bouting" if the calcium set point is caught between a discontinuity,  [(see, e.g., Olypher & Prinz, 2010)](http://dx.doi.org/10.1007/s10827-010-0213-z). This is a degenerate case where the homeostatic mechanism fails. However, there are many regions of conductance space where the time-average calcium function is sufficiently smooth for homeostatic regulation to function properly. An entirely passive model, which we study here, is a very good approximation of these well-behaved systems [(O'Leary et al., 2013](http://dx.doi.org/10.1073/pnas.1309966110); [O'Leary et al., 2014)](http://alexhwilliams.info/pubs/OLeary_etal_2014.pdf).
