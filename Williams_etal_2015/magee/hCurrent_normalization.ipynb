The temporal summation of inputs at CA1 pyramidal somata are remarkably invariant to the location of synaptic input ([Magee, 1999](http://dx.doi.org/10.1038/9158); [Magee, 2000](http://dx.doi.org/10.1038/35044552)). This property of dendritic integration is thought to be mediated by the h current (I<sub>h</sub>), whose magnitude increases along the main apical dendrite of CA1 cells ([Magee, 1998](http://www.jneurosci.org/content/18/19/7613.short)). This notebook shows how to model this effect.

```python
from __future__ import division
from neuron import h
import numpy as np
import pylab as plt
from PyNeuronToolbox.synapses import add_exp2
%matplotlib inline
plt.rcParams['figure.figsize'] = (7.5,10)

# create model
h.load_file('stdrun.hoc')
h.tstop = 20e3
ts = 3*h.tstop/4
tr = h.tstop-ts

Ncomp = 5
sec = []
syndata = []
for i in range(Ncomp):
    sec.append(h.Section())
    sec[-1].nseg = 1
    sec[-1].diam = 1  # um
    sec[-1].L = 100    # um
    sec[-1].Ra = 100
    
    # Leak conductance
    sec[-1].insert('pas')
    sec[-1].g_pas = 0.0003
    sec[-1].e_pas = -85
    
    # h-current
    sec[-1].insert('hd')
    sec[-1].ghdbar_hd = 0.0003 + (i/Ncomp)*1e-2
    
    # Add a synapse in the middle of each segment
    spktimes = np.linspace(ts + tr*((i+0.35)/Ncomp), ts + tr*((i+0.65)/Ncomp), 5)
    syndata.append(add_exp2(h,sec[-1](0.5),spktimes,e=0,tau1=0.5,tau2=20,weight=0.01))
    
    if i > 0: sec[-2].connect(sec[-1],0,1)
    
# Specify variables to record.
# For simplicity, I've put everything in a python dict called "data"
data = {'v':[]}
for s in sec:
    data['v'].append(h.Vector())
    data['v'][-1].record(s(0.5)._ref_v)

# Run simulation
h.run()
t = np.linspace(0,h.tstop,len(data['v'][0])) 
```

```python
# Plot the membrane potential in each compartment

#cols = []
#for i in range(Ncomp): cols.append((1,0,0))
cols = np.vstack((np.linspace(0.5,0,Ncomp),np.linspace(0.5,0,Ncomp),np.linspace(0.5,1,Ncomp))).T

plt.figure(figsize=(5,5))
v_soma = np.array(data['v'][0].to_python())
v_soma_ss = v_soma[t>ts]
t_ss = t[t>ts]
v_soma_ss = np.reshape(v_soma_ss,(Ncomp,len(t_ss)/Ncomp))
it = v_soma_ss.shape[1]
for i in range(Ncomp): plt.plot(t[:it],v_soma_ss[i,:],'-',lw=2,c=cols[i,:])
plt.ylim([-80,0])
plt.xlim([0,t[it]])
plt.yticks(np.arange(-80,-40,20))
plt.ylabel('v (mV)', fontweight='bold')
plt.xlabel('t (ms)', fontweight='bold') # label x-axis on bottom graph
```

```python
# Plot the membrane potential in each compartment
plt.figure()
for i in range(Ncomp):
    plt.subplot(Ncomp,1,i+1)
    
    # voltage trace
    plt.plot(t,data['v'][i],'-b',lw=2)
    
    plt.ylim([-80,60])
    plt.xlim([ts,h.tstop])
    plt.yticks(np.arange(-80,60,50))
    plt.ylabel('v (mV)', fontweight='bold')
    plt.setp(plt.gca().get_xticklabels(), visible=False)
    
    if i == Ncomp-1:
        plt.xlabel('t (ms)', fontweight='bold') # label x-axis on bottom graph
        plt.setp(plt.gca().get_xticklabels(), visible=True)
```
