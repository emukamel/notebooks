```julia
using Distributions
using PyPlot
```

```julia
function plot_meth(X)
    N,W = size(X)
    for n = 1:N
        plot([0.5,W+0.5],[n,n],"-k")
        for w = 1:W
            if X[n,w]
                plot(w,n,"ok")
            else
                plot(w,n,"ow")
            end
        end
    end
    xlim([0,1+W])
    ylim([0,1+N])
    xticks([])
    yticks([])
end
```

```julia
Dparams = [[1 1 1],[10 10 10],[0.1 0.1 0.1],[0.5 1.5 1]]

f = figure()
for s = 1:4
    ax = subplot(2,2,s,projection="3d")
    p = squeeze(Dparams[s,:],[1])
    X = transpose(rand(Dirichlet(p),500));
    plot3D(X[:,1],X[:,2],X[:,3],".")
    ax[:view_init](20,10)
    xlim([0,1])
    ylim([0,1])
    zlim([0,1])
    xticks([0,1])
    yticks([0,1])
    zticks([0,1])
    title(p)
end
```
