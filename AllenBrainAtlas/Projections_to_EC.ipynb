Projections to Lateral vs. Medial Entorhinal Cortex in the Allen Connectivity Atlas
=====

Instructions for downloading .csv files:
* Go to http://connectivity.brain-map.org/
* Click on "Target Search" on the left panel
* Clear the "Filter Source Structure" box on the left
* In the "Target Structure" box, type **ENTl** for *lateral* or **ENTm** for *medial* EC
* Scroll down and click "Download as CSV" in the search results box

```python
from __future__ import division
import numpy as np
import warnings
warnings.filterwarnings('ignore')

def get_data(filename):
    data = {}
    with open(filename,'r') as f:
        lines = f.readlines()
        colheaders = lines[0].split('\t')
        i = colheaders.index('structure-name') # column for name
        j = colheaders.index('sum') # column for fluorescence
        k = colheaders.index('num-voxels')
        m = colheaders.index('injection-volume')
        for line in lines[1:]:
            row = line.split('\t')
            val = float(row[j])/float(row[k])
            if ~np.isnan(val) and val>1e-9:
                if row[i] not in data.keys(): data[row[i]] = {'vals':[],'inj':[]}
                data[row[i]]['vals'].append(np.log(val))
                data[row[i]]['inj'].append(float(row[m]))
                
    return data

LEC = get_data('projections_to_LEC.csv')
MEC = get_data('projections_to_MEC.csv')
```

```python
import pylab as plt
from scipy.stats import ttest_ind,t
%matplotlib inline

# list of common structures
names = list(set(LEC.keys()) & set(MEC.keys()))

def bootstrap_mean_diff(x,y):
    xx = np.random.choice(x,size=len(x),replace=True)
    yy = np.random.choice(y,size=len(y),replace=True)
    if (len(x)+len(y)-2) > 0:
        tstat,pval = ttest_ind(xx,yy)
    else:
        tstat,pval = np.nan,np.nan
    return np.mean(xx)-np.mean(yy),tstat

def est_effect_size(x,y,n=2000):
    estimates,tstats = [],[]
    for i in range(n):
        mean_diff,tstat = bootstrap_mean_diff(x,y)
        estimates.append(mean_diff)
        tstats.append(tstat)
    
    T = np.percentile(tstats,[2.5,97.5])
    df = len(x)+len(y)-2
    pval = min([t.sf(T[0],df),1-t.sf(T[1], df)])
    significance = pval < 0.05
    return np.mean(estimates),significance

N,E,S = [],[],[]
for name in names:
    LEC_data,MEC_data = LEC[name]['vals'],MEC[name]['vals']
    N.append(len(LEC_data) + len(MEC_data))
    if N[-1] < 3:
        effect,significance = 0,False
    else:
        effect,significance = est_effect_size(LEC_data,MEC_data)
    E.append(effect)
    S.append(significance)
```

```python
plt.figure()
plt.hist(E,bins=20)
plt.show()
```

```python
plt.figure(figsize=(10,10))

n_per_plot = 5
n_plots = 2
s,xx = 1,0
plt.subplot(n_plots,1,1)
sig_names = []
for i in np.flipud(np.argsort(E)):
    name = names[i]
    LEC_data,MEC_data = LEC[name]['vals'],MEC[name]['vals']
    if S[i]:
        xx += 1
        if xx > n_per_plot:
            xx = 1
            s += 1
            plt.subplot(n_plots,1,s)
        plt.plot(xx+np.random.uniform(-0.3,-0.1,len(LEC_data)),LEC_data,'or')
        plt.plot(xx+np.random.uniform(0.1,0.3,len(MEC_data)),MEC_data,'ob')
        sig_names.append(name.replace(' ','\n'))
    

for s in range(n_plots):
    plt.subplot(n_plots,1,s+1)
    plt.plot([0,n_per_plot+1],-13*np.ones(2),'--k')
    plt.xlim([0,n_per_plot+1])
    plt.xticks(range(1,n_per_plot+1),sig_names[s*n_per_plot:((s+1)*n_per_plot)],rotation=-45)
    plt.ylim([-22,-8])
    plt.ylabel('normalized fluorescence',fontweight='bold')

plt.legend(['LEC','MEC'],loc='lower right')
plt.tight_layout()
plt.show()
```

The injection volume is significantly correlated with target fluorescence only in a subset of structures
--------------

The experiments in the connectivity atlas use a range of injection volumes, creating a potential confound in the above analysis. To explore this, we plotted the correlation coefficient between the injected volume and normalized target fluorescence between each source structure and the MEC/LEC. Surprisingly, the injection volume was strongly correlated target fluorescence in only a subset of source structures.

```python
from scipy.stats import pearsonr,spearmanr
from matplotlib.pyplot import cm

plt.figure(figsize=(15,6))
s = 1
for corrfunc in pearsonr,spearmanr:
    plt.subplot(1,2,s)
    s += 1
    lec_corr,mec_corr,pvals,nn = [],[],[],[]
    for name in names:
        r1,p1 = corrfunc(LEC[name]['inj'],LEC[name]['vals'])
        r2,p2 = corrfunc(MEC[name]['inj'],MEC[name]['vals'])
        if np.any(np.isnan([p1,p2])): continue
        lec_corr.append(r1)
        mec_corr.append(r2)
        pvals.append(max([p1,p2]))
        nn.append(len(LEC[name]['inj'])+len(MEC[name]['inj']))
    pvals,lec_corr,mec_corr,nn = np.array(pvals),np.array(lec_corr),np.array(mec_corr),np.array(nn)
    plt.scatter(lec_corr,mec_corr,c=pvals,cmap=cm.seismic_r,clim=[-max(pvals)+0.05,max(pvals)+0.05],s=nn+20)
    plt.xlim([-1.2,1.2])
    plt.ylim([-1.2,1.2])
    plt.xlabel('Lateral Entorhinal',fontweight='bold')
    plt.ylabel('Medial Entorhinal',fontweight='bold')

plt.subplot(1,2,1)
plt.title('Pearson Correlation Coefficient',fontweight='bold')
plt.subplot(1,2,2)
plt.title('Spearman Correlation Coefficient',fontweight='bold')
plt.show()
```

**Legend**: The Pearson *(Left)* and Spearman *(Right)* correlation coefficients between injection volume and normalized target fluorescence. The color of each dot represents the significance (p-value) of the calculated correlation coefficient (red, p < 0.05; white, p = 0.05; blue, p > 0.05). The size of each dot denotes the number of replications for each experiment.
