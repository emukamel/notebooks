```python
import zipfile
from StringIO import StringIO
import urllib
from struct import unpack

# Download a dataset (in the future, iterate over many datasets)
pagedUrl = 'http://api.brain-map.org/grid_data/download/126862385?include=density'
myzipfile = zipfile.ZipFile(StringIO(urllib.urlopen(pagedUrl).read()))

# List of files from downloaded dataset
for name in myzipfile.namelist():
    print name
```

```python
## Read the intensity.raw and density.raw files
import numpy as np

# Extract dimensions of 3D grid
with myzipfile.open('density.mhd','r') as f:
    for line in f.readlines():
        if line[:7] == 'DimSize':
            dims = [int(d) for d in line[10:].split(' ')]
dims.reverse()
print dims

# There should be 4 characters '\xhh\xhh\xhh\xhh' for each 32-bit number
with myzipfile.open('density.raw','r') as f:
    lines = f.readlines()
    s = ''.join(lines)  # turn file into one large string. What are the newlines for in the original file?
    density = unpack('<%df'%(len(s)/4),s)

# map -1 to zero
density = np.array(density)
density[density < 0] *= 0
```
