```python
import urllib
import xmltodict # https://github.com/martinblech/xmltodict
#rows = '&start_row=%d&num_rows=%d'%(0,19985) # this would download everything
rows = '&start_row=%d&num_rows=%d'%(0,100)
urlquery = ('http://api.brain-map.org/api/v2/data/Gene/query.xml'
           '?criteria=products%5Bid$eq1%5D'+rows)
print urlquery
s = urllib.urlopen(urlquery).read()
d = xmltodict.parse(s)
genes = d['Response']['objects']['object'] # this is a list of dicts

gene_names = []
for gene in genes:
    gene_names.append(gene['acronym'])
```

```python
# download metadata for a random gene
import random
from collections import OrderedDict
name = random.choice(gene_names)
print name
urlquery = ('http://api.brain-map.org/api/v2/data/SectionDataSet/query.xml'
            '?criteria=products%5Bid$eq1%5D,genes%5Bacronym$eq%27'+name+
            '%27%5D&include=genes,section_images')
print urlquery
s = urllib.urlopen(urlquery).read()
#print s
d = xmltodict.parse(s)

sections = d['Response']['section-data-sets']['section-data-set']
if type(sections) is OrderedDict:
    sections = [sections] # make sections a list

dataset_ids,img_ids = [],[]
for sec in sections:
    dataset_ids.append(sec['id'])
    img_ids.append([])
    for img in sec['section-images']['section-image']: # list of section images
        img_ids[-1].append(img['id']['#text'])
```
