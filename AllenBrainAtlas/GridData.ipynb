```python
import zipfile
from StringIO import StringIO
import urllib
from struct import unpack

# Download a dataset (in the future, iterate over many datasets)
pagedUrl = 'http://api.brain-map.org/grid_data/download/69816930?include=intensity,density'
myzipfile = zipfile.ZipFile(StringIO(urllib.urlopen(pagedUrl).read()))

# List of files from downloaded dataset
for name in myzipfile.namelist():
    print name
```

```python
## Read the intensity.raw and density.raw files
import numpy as np

# There should be 4 characters '\xhh\xhh\xhh\xhh' for each 32-bit number
with myzipfile.open('intensity.raw','r') as f:
    lines = f.readlines()
    s = ''.join(lines)  # turn file into one large string. What are the newlines for in the original file?
    intensity = unpack('<%df'%(len(s)/4),s)

with myzipfile.open('density.raw','r') as f:
    lines = f.readlines()
    s = ''.join(lines)  # turn file into one large string. What are the newlines for in the original file?
    density = unpack('<%df'%(len(s)/4),s)

# map -1 to zero
intensity = np.array(intensity)
intensity[intensity < 0] *= 0
density = np.array(density)
density[density < 0] *= 0
```

**Expression Gridding: ** http://help.brain-map.org/display/mousebrain/API

For each SectionDataSet, the Gridding module creates a low resolution 3-D summary of the gene expression and projects the data to the common coordinate space of the 3-D reference model. Casting all data into a canonical space allows for easy cross-comparison of gene expression data from every Product. The expression data grids can also be viewed directly as 3-D volumes or used for analysis (i.e. differential and correlative searches).

Each image in a SectionDataSet is divided into a 200 x 200 µm grid. Pixel-based gene expression statistics are computed using information from the primary ISH and the expression mask:
