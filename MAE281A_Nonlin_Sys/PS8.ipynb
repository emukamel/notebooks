```python
from __future__ import division
import numpy as np
import pylab as plt
from scipy.integrate import odeint
%matplotlib inline

eps = 0.005
def f(xx, t):
    x,y,z = xx[0],xx[1],xx[2]
    return [eps*(-x+1-2*(y+(np.sin(t)**2))),
            eps*z,
            eps*(-z-np.sin(t)*((x/2)+((y+np.sin(t))**2)))]

t = np.arange(0, 500, 0.1)
xx0 = [2,-1,0]#np.random.uniform(-1,1,3)
xx = odeint(f, xx0, t)

plt.figure()
plt.plot(t,xx)
plt.legend(['x','y','z'],loc='best')
plt.show()
```

```python
def f(xx, t):
    x,y,z = xx[0],xx[1],xx[2]
    return [-2*np.pi*x - 8*np.pi*(y**2),
            z,
            -x - 2*y - 2*z]

t = np.arange(0, 10, 0.1)
xx0 = [0,0,0.1]#np.random.uniform(-1,1,3)
xx = odeint(f, xx0, t)

plt.figure()
plt.plot(t,xx)
plt.legend(['x','y','z'],loc='best')
plt.show()
```
