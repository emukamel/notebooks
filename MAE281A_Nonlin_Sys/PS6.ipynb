```python
from __future__ import division
import numpy as np
import pylab as plt
from scipy.integrate import odeint
%matplotlib inline

def f(x, t):
    x1,x2,x3 = x[0],x[1],x[2]
    return [x2 + x1*x3,
            -x1 - x2 + x2*x3,
            -x1**2 - x2**2]

t = np.arange(0, 10, 0.1)
x0 = np.random.uniform(-1,1,3)
x = odeint(f, x0, t)

plt.figure()
plt.plot(t,x)
plt.legend(['x1','x2','x3'],loc='best')
plt.show()
```
