```python
from __future__ import division
import numpy as np
import pylab as plt
%matplotlib inline
```

**Problem 1 (1)**

```python
## Print eigenvalues of Jacobian at equilibria
J = np.matrix([[0,1],[-1,-1]])
val,vec = np.linalg.eig(J)
print val

J = np.matrix([[0,1],[2,-1]])
val,vec = np.linalg.eig(J)
print val

J = np.matrix([[0,1],[-4,-1]])
val,vec = np.linalg.eig(J)
print val

## Streamplot to check answer
X2, X1 = np.mgrid[-3:3:100j, -3:3:100j]
X1dot = X2
X2dot = -X1 + (1/6)*(X1**3) - X2
speed = np.sqrt(X1dot*X1dot + X2dot*X2dot)

plt.streamplot(X1, X2, X1dot, X2dot, color=speed, linewidth=2, cmap=plt.cm.autumn)
plt.colorbar()
plt.show()
```

```python
## System #4
X2, X1 = np.mgrid[-1.5:1.5:100j, -1.5:1.5:100j]

def phi(y):
    z = np.zeros(y.shape)
    for i in range(y.shape[0]):
        for j in range(y.shape[1]):
            if abs(y[i,j]) <= 1:
                z[i,j] = y[i,j]**3 + 0.5*y[i,j]
            else:
                z[i,j] = 2*y[i,j] - 0.5
    return z

X1dot = X2
X2dot = -X2 - phi(X1-X2)
speed = np.sqrt(X1dot*X1dot + X2dot*X2dot)

plt.streamplot(X1, X2, X1dot, X2dot, color=speed, linewidth=2, cmap=plt.cm.autumn)
plt.colorbar()
plt.xlim([-1.5,1.5])
plt.ylim([-1.5,1.5])
plt.show()
```
