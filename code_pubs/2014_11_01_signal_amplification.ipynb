Averaging Across pixels to Amplify the Signal of Voltage-Indicators
-------------------------------------------------------------------

This is a simple notebook that shows how very effectively extract a "theta rhythm" from what appears to be pure noise. See this [wikipedia article](http://en.wikipedia.org/wiki/Signal_averaging) for more details. This explains that the signal to noise ratio should (ideally) be increased in proportion to the square root of the number averaged measurements (in this case, number of pixels).

```python
from __future__ import division
import numpy as np
import pylab as plt
%matplotlib inline
```

```python
# A synthetic theta rhythm
sampfreq = 1e4  # Hz
tmax = 10       # seconds 
freq = 8        # Hz of theta

nt = sampfreq*tmax  # number of samples over time
t = np.linspace(0,tmax,nt)
y = np.sin(t*freq*2*np.pi)

plt.figure()
plt.plot(t[t<1],y[t<1])
plt.ylabel('signal',fontweight='bold',fontsize=15)
plt.xlabel('time (s)',fontweight='bold',fontsize=15)
plt.show()
```

```python
# repeat y across N pixels  (40x40 pixel image as example)
N = 40**2
y = np.ones((1,nt))*y
y = np.dot(np.ones((N,1)),y).T
```

```python
plt.figure()
plt.plot(t[t<1],y[t<1,0],'-b')
plt.plot(t[t<1],y[t<1,99],'-r')
plt.ylabel('signal',fontweight='bold',fontsize=15)
plt.xlabel('time (s)',fontweight='bold',fontsize=15)
plt.show()
```

```python
# Add an insane amount of noise
noise = np.random.normal(scale=50.0,size=y.shape)
yn = y+noise

plt.figure()
plt.plot(t[t<1],yn[t<1,0],'-b')
plt.ylabel('noisy signal',fontweight='bold',fontsize=15)
plt.xlabel('time (s)',fontweight='bold',fontsize=15)
plt.show()
```

```python
# Print avearge signal to noise ratio for each pixel:
print 'S/N  =  ' + \
      str(np.mean(np.var(y,axis=0))/np.mean(np.var(noise,axis=0))) + \
      ' (before averaging)'
print 'S/N  =  ' + \
      str(np.var(np.mean(y,axis=1))/np.var(np.mean(noise,axis=1))) + \
      ' (after averaging)'
```

```python
# Make a movie to show synthetic dataset
from matplotlib import animation
from tempfile import NamedTemporaryFile
from IPython.display import HTML

climits = [np.min(yn),np.max(yn)]

nframes = 100
fig = plt.figure(figsize=(8,8))

def anim_func(i):
    ti = int(round((i/nframes)*(sampfreq/freq)))
    plt.pcolor(yn[ti,:].reshape(np.sqrt(N),np.sqrt(N)),cmap=plt.get_cmap('Reds'),clim=climits)

VIDEO_TAG = """<video controls>
 <source src="data:video/x-m4v;base64,{0}" type="video/mp4">
 Your browser does not support the video tag.
</video>"""

def anim_to_html(anim):
    if not hasattr(anim, '_encoded_video'):
        with NamedTemporaryFile(suffix='.mp4') as f:
            anim.save(f.name, fps=20, extra_args=['-vcodec', 'libx264'])
            video = open(f.name, "rb").read()
        anim._encoded_video = video.encode("base64")
    
    return VIDEO_TAG.format(anim._encoded_video)

animation.Animation._repr_html_ = anim_to_html

# Make and show an animation
anim = animation.FuncAnimation(fig, anim_func, frames=nframes, interval=60, blit=True)
HTML(anim_to_html(anim))
```

```python
# Try finding the theta in a signal pixel (hopeless)
from scipy.signal import butter, filtfilt

nyqfreq = sampfreq/2.0
F = np.linspace(0,nyqfreq,nt)
Wn = []
Wn.append(np.argmin(np.abs(F-6))/len(F))  # Low cutoff at 6Hz
Wn.append(np.argmin(np.abs(F-10))/len(F)) # High cutoff at 10Hz
b,a = butter(1, Wn, btype='bandpass')

yf = filtfilt(b, a, yn[:,0])
nf = filtfilt(b, a, noise[:,0])

plt.figure(figsize=(8,6))
ti = (t > 1) & (t < 2)

plt.subplot(2,2,1)
plt.plot(t[ti],yn[ti,0])
plt.title('Noisy Trace, with theta',fontweight='bold',fontsize=15)

plt.subplot(2,2,2)
plt.plot(t[ti],noise[ti,0])
plt.title('Noisy Trace, no theta',fontweight='bold',fontsize=15)

plt.subplot(2,2,3)
plt.plot(t[ti],y[ti,0],lw=2,color=(0.3,0.3,1.0))
plt.plot(t[ti],yf[ti],'-r',lw=1.5)
plt.title('Reconstructed (red)\n target (blue)',fontweight='bold',fontsize=15)

plt.subplot(2,2,4)
plt.plot(t[ti],nf[ti],'-r',lw=1.5)
plt.title('Reconstructed',fontweight='bold',fontsize=15)

plt.tight_layout()
plt.show()
```

```python
# Average and try filter again
yavg = np.mean(yn,axis=1)
noiseavg = np.mean(noise,axis=1)

yf = filtfilt(b, a, yavg)
nf = filtfilt(b, a, noiseavg)

plt.figure(figsize=(8,6))
ti = (t > 1) & (t < 2)

plt.subplot(2,2,1)
plt.plot(t[ti],yavg[ti])
plt.title('Noisy Trace, with theta',fontweight='bold',fontsize=15)

plt.subplot(2,2,2)
plt.plot(t[ti],noiseavg[ti])
plt.title('Noisy Trace, no theta',fontweight='bold',fontsize=15)

plt.subplot(2,2,3)
plt.plot(t[ti],y[ti,0],lw=2,color=(0.3,0.3,1.0))
plt.plot(t[ti],yf[ti],'-r',lw=1.5)
plt.title('Reconstructed (red)\n target (blue)',fontweight='bold',fontsize=15)

plt.subplot(2,2,4)
plt.plot(t[ti],nf[ti],'-r',lw=1.5)
plt.title('Reconstructed',fontweight='bold',fontsize=15)

plt.tight_layout()
plt.show()
```
