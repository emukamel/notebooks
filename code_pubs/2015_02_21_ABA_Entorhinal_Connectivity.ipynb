**API documentation for downloading Grid Data:** http://help.brain-map.org/display/api/Downloading+3-D+Expression+Grid+Data

This notebook downloads all connectivity experiments originating from the lateral and the medial entorhinal cortex, and compares their averaged projection patterns.

```python
# Import experimental ids
def get_ids(filename):
    ids = []
    with open(filename,'r') as f:
        lines = f.readlines()
        i = lines[0].split(',').index('id') # column for ids
        for line in lines[1:]:
            ids.append(line.split(',')[0])    
    return ids

medial_ids = get_ids('medial.csv')
lateral_ids = get_ids('lateral.csv')
```

```python
import zipfile
from StringIO import StringIO
import urllib
from struct import unpack
import numpy as np

def get_data(expID):
    print 'fetching.... '+expID
    # Download a dataset (in the future, iterate over many datasets)
    pagedUrl = 'http://api.brain-map.org/grid_data/download/'+expID+'?include=density'
    myzipfile = zipfile.ZipFile(StringIO(urllib.urlopen(pagedUrl).read()))

    # Extract dimensions of 3D grid
    with myzipfile.open('density.mhd','r') as f:
        for line in f.readlines():
            if line[:7] == 'DimSize':
                dims = [int(d) for d in line[10:].split(' ')]
    dims.reverse()

    ## Read the density.raw file
    # There should be 4 characters '\xhh\xhh\xhh\xhh' for each 32-bit number
    with myzipfile.open('density.raw','r') as f:
        lines = f.readlines()
        s = ''.join(lines)  # turn file into one large string. What are the newlines for in the original file?
        density = unpack('<%df'%(len(s)/4),s)

    # map -1 to zero
    density = np.array(density)
    density[density < 0] *= 0

    # Make 3D arrays of intensity, density, and expression
    D = np.reshape(density,tuple(dims))

    return D
```

```python
import pylab as plt
from matplotlib.pyplot import cm
%matplotlib inline

print 'fetching medial datasets....'
avg_medial = get_data(medial_ids[0])
for expid in medial_ids[1:]:
    avg_medial += get_data(expid)
avg_medial /= len(medial_ids)

print 'fetching lateral datasets....'
avg_lateral = get_data(lateral_ids[0])
for expid in lateral_ids[1:]:
    avg_lateral += get_data(expid)
avg_lateral /= len(lateral_ids)
```

```python
# Compare averaged projection patterns
plt.figure(figsize=(10,10))
clim = np.percentile(avg_medial.flatten(),[1,99])
for i in range(3):
    ax = plt.subplot(3,2,i*2+1)
    max_projection = np.max(avg_medial,axis=i)
    if i == 2: max_projection = max_projection.T
    ax.imshow(max_projection,cmap=cm.cubehelix,clim=clim)
    if i == 0: plt.title('medial',fontweight='bold')

clim = np.percentile(avg_lateral.flatten(),[1,99])
for i in range(3):
    ax = plt.subplot(3,2,i*2+2)
    max_projection = np.max(avg_lateral,axis=i)
    if i == 2: max_projection = max_projection.T
    ax.imshow(max_projection,cmap=cm.cubehelix,clim=clim)
    if i == 0: plt.title('lateral',fontweight='bold')

plt.savefig('avg.png')
```

```python
# Subtract the left plots from the right plots above to highlight differences
avg_diff = avg_medial-avg_lateral
clim = np.percentile(avg_diff.flatten(),[1,99])
clim = [-max(abs(clim)), max(abs(clim))]

plt.figure(figsize=(10,10))
for i in range(3):
    ax = plt.subplot(3,1,i+1)
    projection = np.mean(avg_diff,axis=i)
    if i == 2: projection = projection.T
    ax.imshow(projection,cmap=cm.seismic,clim=clim)
    if i == 0: plt.title('Red (medial > lateral)\nBlue (lateral > medial)',fontweight='bold')

plt.savefig('avg_diff.png')
```
