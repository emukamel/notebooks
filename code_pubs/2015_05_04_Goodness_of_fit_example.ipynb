```python
import numpy as np
import pylab as plt
from scipy import stats
%matplotlib inline
```

```python
X = np.random.normal(0,1,200)
x1 = X[:100]
x2 = X[100:]
y1 = x1 + np.random.normal(0,0.2,100)
y2 = 2*x2 + np.random.normal(0,0.2,100)
Y = np.concatenate((y1,y2))
slope, intercept, r_value, p_value, std_err = stats.linregress(X,Y)

plt.figure()
plt.plot(np.linspace(-3,3),np.linspace(-3,3)*slope + intercept,'-k',lw=2)
plt.plot(x1,y1,'.b')
plt.plot(x2,y2,'.r')
plt.xlim([-3,3])
plt.ylim([-6,6])
plt.legend(['best fit','peptide 1','peptide 2'],loc='best')
plt.show()
```

```python
y1_predicted = x1*slope + intercept
y2_predicted = x2*slope + intercept

y1_resid = y1 - y1_predicted
y2_resid = y2 - y2_predicted

plt.figure()
plt.hist(y1_resid,color='b')
plt.hist(y2_resid,color='r',alpha=0.5)
plt.xlabel('residual',fontweight='bold')
plt.title('residuals not significantly different',fontweight='bold')
plt.show()
```
